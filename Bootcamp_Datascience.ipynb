# **Aula 1**

**Import da biblioteca Panda**

```
import pandas as pd
```

**Set-up de como aparecerão os dados (linhas, colunas, separadores)**

```
fonte = '/content/A151346189_28_143_208.csv'

dados = pd.read_csv(fonte, encoding='latin1', skiprows =3, sep = ";", skipfooter=12, thousands=".", decimal=",")
dados
```

**Formatação dos dados com casas decimais**

```
  pd.options.display.float_format = "{:.2f}".format
```

**dados.head() os 5 primeiros dados e "tail", os 5 finais**

```
dados.head()
```

**Média de todas as colunas**

```
dados.mean() #média
```

**Média da coluna 2008/Ago**


```
dados['2008/Ago']
```

**Média dos dados da coluna 2008/Ago com mean**


```
dados['2008/Ago'].mean()
```

**Primeiro plot de dados, com ativação das bibliotecas pyplot para os gráficos e ticker para alteração nos eixos**

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação",y="2008/Ago",kind="bar",figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor por unidade da federação")
plt.show()
```

**Substituição do object por 0, somente no collab com "inplace"**


```
dados.replace('-', 0, inplace=True)
```

**Após a substituição, modificação da coluna para float**

```
dados['2020/Ago'] = dados['2020/Ago'].astype(float)
```

**Em seguida, o desafio 01**

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação",y="2020/Ago",kind="bar",figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R${x:,.2f}"))
plt.title("Gastos hospitalares por unidade da federação - Ago/2020", fontsize=16)
plt.show()
```

**Exercício do ordenar os dados**


```
ax = dados.sort_values(by='2020/Ago', ascending=False).plot(x='Unidade da Federação',y='2020/Ago',kind='bar',figsize=(12,5))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R${x:,.2f}"))

```

# **Aula 2**

```
fonte = '/content/A160324189_28_143_208.csv' 

dados = pd.read_csv(fonte, encoding='latin1', skiprows =3, sep = ";", skipfooter=12, thousands=".", decimal=",")
dados
```

```
  pd.options.display.float_format = "{:.2f}".format
```

**Para trazer mais de uma colunas, é preciso trazer 2 colchetes**

```
dados[['2008/Ago','2008/Set']]
```

**Para visualizar as colunas por nome, mean + index**

```
dados.mean().index
```

**Renomeando o index e inseri a coluna unidade da federação**

```
colunas_usaveis = dados.mean().index.tolist() #index é imutável, com tolist foi possivel add a coluna "unidade"
colunas_usaveis.insert(0,"Unidade da Federação")
colunas_usaveis
```

**Abaixo, printamos somente que nomeei e com dados.**


```
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```
usaveis =  usaveis.set_index("Unidade da Federação")
usaveis
```

**loc é usado para printar linhas**

```
usaveis.loc["12 Acre"]
```

**T transpoe os dados**



```
usaveis.T
```

** drop retira a linha do eixo 1 (col) ou 0 (linha)**

```
usaveis = usaveis.drop("Total", axis=1)
usaveis.head()
```

```
  pd.options.display.float_format = "{:.2f}".format
```

**Segundo desafio: Um melhor local para a legenda e selecionar 5 estados**

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = usaveis.T.plot(y=["12 Acre","13 Amazonas","15 Pará","29 Bahia","52 Goiás"],figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R${x:,.2f}"))
ax.set_ylabel("Investimento", color="Black", fontweight="bold", fontsize=14) # ajuste eixo y
ax.set_xlabel("Período", color="Black", fontweight="bold", fontsize=14) # ajuste eixo x
plt.title("SUS | Gastos hospitalares por unidade da federação - Ago/07 - Ago/20", fontsize=16)
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1)) # Legenda fora do box com bblox
plt.show()
```
